```python
import numpy as np
import matplotlib.pyplot as plt

def hypothesis(X, theta):
    return X.dot(theta)

def cost_function(loss):
    m = len(loss)
    return  np.sum(loss ** 2) / (2 * m)

def gradientDescent(x, y, theta, alpha, m, numIterations):
    J_hist = []

    for i in range(0, numIterations):
        loss = hypothesis(x, theta) - y
        J_hist.append(cost_function(loss))

        gradient = np.dot(x.T, loss) / m
        # update theta
        theta -= alpha * gradient

    return theta, J_hist
```

```python
data = np.genfromtxt('data/data1.csv', delimiter=",")
X, y = data[:, 0], data[:, 1]
```

```python
m, n = data.shape
XMat = np.ones([m, 2])
yMat = y

for i in range(m):
    XMat[i, 1] = X[i]
```

```python
numIterations= 10000
alpha = 0.0005
theta = np.ones(n)
theta, J_hist = gradientDescent(XMat, yMat, theta, alpha, XMat.shape[0], numIterations)
print(theta)
```

```python
plt.scatter(XMat[:,1], yMat, marker="+", c="r")
plt.plot(
    range(100), 
    theta[1]*range(100) + theta[0]
)
```
